## 4.4 Plot the real Pareto front from NSGA-III

Unlike the illustrative animations above, this plot is generated from the actual optimisation results produced in `03_optimisation.ipynb`.

```python
import matplotlib.pyplot as plt
import pandas as pd

pareto_csv = Path('data/optimized/pareto_solutions.csv')
if pareto_csv.exists():
    df = pd.read_csv(pareto_csv)
    # Restore maximisation direction for geometric_accuracy
    if 'geometric_accuracy' in df.columns:
        df['geometric_accuracy'] = -df['geometric_accuracy']

    fig = plt.figure(figsize=(10, 8))
    ax = fig.add_subplot(111, projection='3d')
    scat = ax.scatter(
        df['residual_stress'],
        df['porosity'],
        df['geometric_accuracy'],
        c=df['geometric_accuracy'],
        cmap='viridis',
        s=60,
        edgecolors='k',
        linewidth=0.5,
    )
    ax.set_xlabel('Residual Stress (MPa)')
    ax.set_ylabel('Porosity')
    ax.set_zlabel('Geometric Accuracy')
    ax.set_title('Pareto Front from NSGA-III Surrogate Optimisation')
    fig.colorbar(scat, ax=ax, label='Geometric Accuracy')
    plt.tight_layout()
    plt.savefig('docs/pareto_front_notebook.png', dpi=200)
    plt.show()
else:
    print(f"{pareto_csv} not found. Run 03_optimisation.ipynb first.")
```
